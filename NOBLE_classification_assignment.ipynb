```python
# filling missing values

for i in ['SL','SW','PL','PW']:
  df[i]=df[i].fillna(df[i].mean())
df.isnull().sum()
```

```python
df['Classification'].nunique()
```

```python
# encoding catdegorical variables

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Classification'] = le.fit_transform(df['Classification'])
df
```

```python
df['Classification'].value_counts()
```

```python
df.describe()
```

```python
#standerd scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df[['SL','SW','PL','PW']] = sc.fit_transform(df[['SL','SW','PL','PW']])
df
```

```python
# split the data into features and labels

x = df.drop('Classification',axis=1)
y = df['Classification']

# split the data into training and testing sets

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**MODEL SELECTION AND TRAINING**


LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logit_model = LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred = logit_model.predict(x_test)


from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))




```

```python
confusion_matrix(y_test,y_pred)
```

K- NEAREST NEIGHBOR

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k = []
neighbors  = np.arange(3,15)
for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(x_train,y_train)
  y_pred = classifier.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  metric_k.append(acc)
```

```python
metric_k
```

```python
classifier = KNeighborsClassifier(n_neighbors=5, metric= 'minkowski', p=2 )
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
acc = accuracy_score(y_test,y_pred)

print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))
```

SVM


```python
from sklearn.svm import SVC
sv_clf = SVC()
sv_clf.fit(x_train,y_train)
```

```python
y_pred = sv_clf.predict(x_test)

print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(random_state= 42)
dt_clf.fit(x_train,y_train)
```

```python
y_pred = dt_clf.predict(x_test)

print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))
```

RANDOM FOREST DECISION TREE

```python

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(x_train,y_train)
```

```python
y_pred = rf_clf.predict(x_test)

print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))
```

HYPER PARAMETER TUNING

```python
# HYPER PARAMETER TUNING

rf_clf1 = RandomForestClassifier(criterion='Entropy', n_estimators=20)

rf_clf = RandomForestClassifier()
rf_clf.fit(x_train,y_train)

y_pred = rf_clf.predict(x_test)


print("accuracy is ", accuracy_score(y_test,y_pred))
print("precision is ", precision_score(y_test,y_pred,average='micro'))
print("recall is ", recall_score(y_test,y_pred,average='micro'))
print("f1 score is ", f1_score(y_test,y_pred,average='micro'))
```

# **The Random Forest model is preferred because, despite achieving the same perfect accuracy as the Decision Tree, it is less likely to overfit the data due to its ensemble nature, providing greater robustness and generalization to unseen data.
